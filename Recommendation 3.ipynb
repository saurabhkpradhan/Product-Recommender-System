```python
import pandas as pd 
df = pd.read_csv("reviews.csv")
df.info()
```

```python
len(pd.unique(df['asin']))
```

```python
len(pd.unique(df['reviewerID']))
```

```python
df2 = df[['reviewerID','asin','sentiment']]
df2
```

```python
df2['sentiment'] = df['sentiment'].replace(['negative','positive'],[0,1])
df2
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df2['asincode'] = lb_make.fit_transform(df2['asin'])
df2.head(20)
```

```python
lb_make = LabelEncoder()
df2['reviewercode'] = lb_make.fit_transform(df2['reviewerID'])
df2.head(20)
```

```python
data = df2[['sentiment','asincode','reviewercode']]
```

```python
df3 = data.groupby('asincode')["sentiment"].count()
df3 = pd.DataFrame(df3)
s = df3['sentiment'].mean()

df3 = df3[df3.sentiment>=s]
l = list(df3.index) 
data = data[data['asincode'].isin(l)]
data.reset_index(drop=True, inplace=True)
data 
```

```python
productMatrix = data.pivot_table(index='asincode',columns = 'reviewercode',values = 'sentiment')
productMatrix
```

```python
productMatrix.fillna(0, inplace=True)
productMatrix.head(10)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim_df = pd.DataFrame(cosine_similarity(productMatrix, productMatrix), index=productMatrix.index, columns=productMatrix.index)
item_sim_df
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
sn.heatmap(item_sim_df)
plt.show()
```

```python
item_sim_df.index
for i in item_sim_df.index:
    l=list(item_sim_df[i].nlargest(n=2))
    if l[1]>=0.6:
        print(item_sim_df[i].nlargest(n=2),end="\n")
```
